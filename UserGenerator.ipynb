```python
import pandas as pd
import names
from random_username.generate import generate_username

import random
from random import randint
import numpy as np
import matplotlib.pyplot as plt
 


def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd ) * np.exp(-2.5*((x-mean)/sd)**2 )
    return prob_density
 


def UsersGenerator(size):
    
    
    x = np.linspace(13,90,size)

    mean = np.mean(x)-20
    sd = np.std(x)
 
    pdf = normal_dist(x,mean,sd)  

    categories =['sport', 'animals', 'nature', 'chair', 'lion', 'example', 'table', 'deer', 'bee', 'sheep', 'fish', 'turkey', 'dove', 'chicken', 'horse']

    column_names = ["name", "username", "age", "interests", "friends","activity","task"]

    df = pd.DataFrame(columns = column_names)
    
        
    for index in range(size):
        
        random.shuffle(categories)
        respondes = randint(0, 5)
        listOfTasks = []
        for i in range(respondes):
            listOfTasks.append(set([random.randint(0,10) for i in range(respondes)]))

        df.at[index, 'name'] = names.get_full_name()
        df.at[index, 'username'] = generate_username(1)[0]
        df.at[index, 'age'] =  int(random.choices(x, pdf)[0])
        df.at[index, 'interests'] = categories[0:randint(0, 10)]
        df.at[index, 'friends'] = set([random.randint(0,size) for i in range(randint(0, 10))])
        df.at[index, 'activity']  = set([random.randint(0,size) for i in range(respondes)])
        df.at[index, 'task']  = listOfTasks
        
    return df


data = UsersGenerator(1000)
data

```
